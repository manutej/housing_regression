```python
## IMPORT LIBRARIES

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./datasets/train_cleaned.csv') # Read in data that was already cleaned
df_orig = pd.read_csv('./datasets/train.csv') # Also read in some of the original data for baseline modeling
df_orig.columns =[col.replace(' ', '_').lower() for col in df_orig.columns] #Replace spaces with columns

df.drop(columns = "Unnamed: 0",inplace = True) # Drop unecessary column from CSV import
df.head()
```

```python
def plot_mlr_opt(df,name):
    ### Instantiate arrays that will aggregate scores in our for loop
    attempts = []
    mse_array_train = []
    mse_array_test = []
    
    rmse_array_train = []
    rmse_array_test = []
    
    r_score_array_train = []
    r_score_array_test = []
    for var in range(2,len(df.columns)): # loop through columns 
        # Our features will automatically be selected here based on correlation to sale price for baselining
        features = list(df.corr()[['saleprice']].sort_values('saleprice',ascending=False).head(var).index)
        # Automatically grabbing the top most correlated with features (with correlation > 0.5)
        df_subset = df.loc[:,features]
        df_subset.fillna(df_subset.median(),inplace = True) #Fill any nulls in data if they exist

        features = list(df_subset.columns)[1:]
        
        target = ['saleprice']
        X = df_subset[features]
        y=df_subset['saleprice']
        X.shape
        X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42) #Test train split with X,y 

        lr = LinearRegression()

        lr.fit(X_train,y_train) # Fit linear regression model
        score_train = lr.score(X_train,y_train) #Training score
        score_test = lr.score(X_test,y_test) # Test Score
        r_score_array_train.append(score_train) # Append score to array for plotting
        r_score_array_test.append(score_test)# Append score to array for plotting
        y_pred = lr.predict(X_train) # Regression model predictions
        y_pred_test = lr.predict(X_test) #Test predictions
        mse_train = mean_squared_error(y_train,y_pred) #Mean squared error for training
        mse_test = mean_squared_error(y_test,y_pred_test) #Mean squared error for testing
        
        rmse_train = mean_squared_error(y_train,y_pred,squared=False) #Root Mean squared error for training
        rmse_test = mean_squared_error(y_test,y_pred_test,squared=False)#Root Mean squared error for testing
        
        ## Add MSE and RMSE to respective aggregation arrays for plotting
        mse_array_train.append(mse_train) 
        mse_array_test.append(mse_test)
        rmse_array_train.append(rmse_train)
        rmse_array_test.append(rmse_test)
        attempts.append(var)
    
    #Plotting
    plt.figure(1,figsize=(10,6))
    plt.scatter(x=attempts,y=r_score_array_train,color='r') #Red is Training score R2
    plt.scatter(x=attempts,y=r_score_array_test,color='b') #Blue is Testing score R2
    plt.legend(['Train',"Test"],fontsize=18)
    plt.xlabel("# of Features Sorted by Connection to Price",fontsize=18)
    plt.ylabel("R Squared Score",fontsize=18)
    plt.axhline(y=max(r_score_array_test),color='k')
    plt.axvline(x=attempts[r_score_array_test.index(max(r_score_array_test))],color='k') #Use to find local max
    
    plt.savefig(f'images/ml_feature_optimization_{name}_r2')
    
    plt.figure(2,figsize=(10,6))
    plt.scatter(x=attempts,y=rmse_array_train,color='r') #Red is Training score RMSE
    plt.scatter(x=attempts,y=rmse_array_test,color='b') #Blue is Testing score RMSE
    plt.legend(['Train',"Test"],fontsize=20)
    
    plt.title("RMSE Squared Results vs Price-Correlated Features",fontsize=20)
    plt.xlabel("# of Features Sorted by Connection to Price",fontsize=18)
    plt.ylabel("RMSE (Root Mean Squared Error) Score",fontsize=18)
    plt.legend(['Train',"Test"],fontsize=18)
    plt.savefig(f'images/ml_feature_optimization_{name}_rmse')
    
    plt.figure(3,figsize=(10,6))
    plt.scatter(x=attempts,y=mse_array_train,color='r') #Red is Training score MSE
    plt.scatter(x=attempts,y=mse_array_test,color='b') #Blue is Testing score MSE
    plt.title("MSE Squared Results vs Price-Correlated Features",fontsize=20)
    plt.xlabel("# of Features Sorted by Connection to Price",fontsize=18)
    plt.ylabel("Mean Squared Error (MSE) Score",fontsize=18)
    plt.legend(['Train',"Test"],fontsize=18)
    plt.savefig(f'images/ml_feature_optimization_{name}_mse')


```

```python
plot_mlr_opt(df,'dummy') # With our features containing dummy variables
```

```python
plot_mlr_opt(df_orig,'original') # With our original data frame
```

### These outputs tell us that there is a limit to how well the model can perform as we add more features. 
This may be a limit imposed by the data or by the conditions of our model. Here in this set of graphs we notice how there is also a more clear asymptote reached for each score as we progressively add more features. Of course, it makes sense that this would happen, because not all features are relevant and at some point, we are adding more "features" to the model, but the quality of the data for those columns may still be such that, we cannot gain anymore improvement in the model. This is a good way to determine our theoretical limits for our models. For our R2 scare we see that this limit is at about 0.85-0.9. 

Another peculiar thing to note here is that our testing set seemed to fare consistently better in the scoring than our training set. 

```python
y = df['saleprice']
X = df.drop(columns = 'saleprice')
```

```python


def get_mlr_regression(X,y):
    
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42)
    ols = LinearRegression()
    
    features = list(df.corr()[['saleprice']].sort_values('saleprice',ascending=False).head(20).index)
    # Automatically grabbing the top most correlated with features (with correlation > 0.5)
    df_subset = df.loc[:,features]
    df_subset.fillna(df_subset.median(),inplace = True)

    features = list(df_subset.columns)[1:]
    
    ols.fit(X_train,y_train)
    score_ols_train = ols.score(X_train,y_train)
    score_ols_test = ols.score(X_test,y_test)
    print(" Regression ".center(18, "="))

    print(f'The R Squared Train score is {score_ols_train}')
    print(f'The R Squared Test score is {score_ols_test}')
    
    
    
    return ols,X_train,X_test,y_train,y_test

ols,X_train_lr,X_test_lr,y_train_lr,y_test_lr = get_mlr_regression(X,y)


```

```python
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import Lasso, LassoCV


def get_split_scale (X,y):
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42)

    sc = StandardScaler()
    Z_train = sc.fit_transform(X_train) # Fit to the means in order to transform
    Z_test = sc.transform(X_test) #Apply the transform to actually use that scaled data on test 
    return Z_train,y_train,Z_test,y_test

Z_train,y_train,Z_test,y_test = get_split_scale (X,y)

def get_ridge_fit(X,y):
    Z_train,y_train,Z_test,y_test = get_split_scale (X,y)
    
    alphas = np.logspace(0,5,100)
    ridge_cv = RidgeCV(alphas = alphas,cv=5)
    ridge_cv.fit(Z_train,y_train);

    return ridge_cv 


def get_lasso_fit(X,y):
    Z_train,y_train,Z_test,y_test = get_split_scale (X,y)
    l_alphas = np.logspace(-3,0,100)
    lasso = LassoCV(alphas = l_alphas,cv=5)
    lasso.fit(Z_train,y_train);
    
    return lasso

ridge_cv = get_ridge_fit(X,y)
lasso = get_lasso_fit(X,y)
print(" Ridge ".center(18, "="))
print('Training Score:')
print(ridge_cv.score(Z_train, y_train))
print('Testing Score:')
print(ridge_cv.score(Z_test, y_test))
print('Optimal Ridge Regression Alpha Co-efficient:')
print(ridge_cv.alpha_)
```

```python
lasso_coefs = pd.Series(lasso.coef_,index = list(X.columns))

plt.figure(figsize=(8, 10))
lasso_coefs[lasso_coefs != 0].sort_values().plot.barh()
plt.title('Nonzero Lasso Coefficients in Relation to Feature Set',fontsize = 18);
plt.xlabel('Lasso Coefficient',fontsize = 16);
plt.ylabel('Feature Name',fontsize = 16);
plt.tight_layout()

plt.savefig('images/lasso_coefficients_barh')
```

### Here we have a plot of our lasso coefficients, showing which coefficients have non-zero values, which means they showed significant relationships in our data in our regularization model. 

```python
ols,X_train_lr,X_test_lr,y_train_lr,y_test_lr = get_mlr_regression(X,y)

print(" Ridge ".center(18, "="))
print(f'The RIDGE Train score is: {ridge_cv.score(Z_train, y_train)}')
print(f'The RIDGE Test score is: {ridge_cv.score(Z_test, y_test)}')
print(" Lasso ".center(18, "="))
print(f'The LASSO Train score is: {lasso.score(Z_train, y_train)}')
print(f'The LASSO Test score is: {lasso.score(Z_test, y_test)}')
```

```python
y_pred_lr = ols.predict(X_train_lr)
y_pred_ridge = ridge_cv.predict(Z_train)
y_pred_lasso = lasso.predict(Z_train)

mse_lr = mean_squared_error(y_train_lr,y_pred_lr)
rmse_lr = mean_squared_error(y_train_lr,y_pred_lr,squared = False)

mse_ridge = mean_squared_error(y_train,y_pred_ridge)
rmse_ridge = mean_squared_error(y_train,y_pred_ridge,squared = False)

mse_lasso = mean_squared_error(y_train,y_pred_lasso)
rmse_lasso = mean_squared_error(y_train,y_pred_lasso,squared = False)

print(" Linear Regression ".center(18, "="))
print(f'Mean Squared error for this model is: {mse_lr}')
print(f'Root Mean Squared error for this model is: {rmse_lr}')

print(" Ridge ".center(18, "="))
print(f'Mean Squared error for this model is: {mse_ridge}')
print(f'Root Mean Squared error for this model is: {rmse_ridge}')
print(" Lasso ".center(18, "="))
print(f'Mean Squared error for this model is: {mse_lasso}')
print(f'Root Mean Squared error for this model is: {rmse_lasso}')
```

## Model Evaluation

Our central assumptions in a multiple linear regression model are that the model follows acronym: "LINEM"
L = Linearity (more relevant for Single Linear Regression (SLR) vs Multiple Linear Regression (MLR)
I = Independence of Observations and Errors
N = Normality of Residuals (Residuals are magnitude of error difference between predicted and true values)
E = Equal variance of errors for 
M = Multiple Co-linearity (relevant for multiple linear regression)


### L = Linearity

```python
plt.figure(figsize = (8,10))
sns.heatmap(df.corr()[['saleprice']].sort_values('saleprice',ascending = False),
           vmin = -1,vmax = 1,cmap = 'coolwarm')
plt.savefig('images/saleprice_heatmap_features_linearity')
```

### I = Independence of Variables (Observations)
We can explain this assumption verbally, because each of our housing data contains distinct rows of distinct houses, and although they may be a relationship between those variables, we can safely gather that they are independent observations, with independent errors. 

### N = Normality of Residuals

```python
def get_residuals(y,y_pred):
    return y - y_pred

def get_residuals_plots(residuals,name):
    plt.figure(figsize = (10,8))
    sns.histplot(residuals,bins = 200,kde = True)
    plt.axvline(x = residuals.mean(),color = 'r')
    plt.title(f'Plot of residiuals for {name} Regression',fontsize = 20)
    plt.xlabel(f'Magnitude of Residual',fontsize = 18)
    plt.ylabel(f'Frequency Count',fontsize = 18)
    plt.tight_layout()
    plt.savefig(f'images/residuals_plot_{name}')
    

residuals_lr = get_residuals(y_train_lr,y_pred_lr)
get_residuals_plots(residuals_lr,'Linear')

residuals_ridge = get_residuals(y_train,y_pred_ridge)
get_residuals_plots(residuals_ridge,'Ridge')

residuals_lasso = get_residuals(y_train,y_pred_lasso)
get_residuals_plots(residuals_lasso,'Lasso')

print('Mean of Residuals for:')
print(f'Linear Regression: {residuals_lr.mean()}')
print(f'Ridge: {residuals_ridge.mean()}')
print(f'Ridge: {residuals_lasso.mean()}')
```

#### We can conclude from these graphs that the magnitude of the residuals for all 3 models follows a relatively normal distribution with a good degree of symmetry and no skew, with a center at 0. 

## E = Equal Variances

```python
def get_variances_plots(residuals,predictions,name):
    plt.figure(figsize = (10,8))
    plt.scatter(y = residuals,x = predictions)
    plt.axhline(y = residuals.mean(),color = 'r')
    plt.title(f'Plot of residiuals vs {name} Predictions',fontsize = 20)
    plt.xlabel(f'{name} Predictions for Sale Price',fontsize = 18)
    plt.ylabel(f'Magnitude of Residual',fontsize = 18)
```

```python
get_variances_plots(residuals_lr,y_pred_lr,'Linear')
get_variances_plots(residuals_ridge,y_pred_ridge,'Ridge')
get_variances_plots(residuals_lasso,y_pred_lasso,'Lasso')
```

### We can see that all of our models do not meet the criteria for equal variance of errors. 

The reason we can make this conclusion is because the scatterplot in all three models fans out more at higher values of our target variable (sale price). Therefore, we can recognize that there is not an equal variance of errors for these three models. We can see from the histogram below that there is a non-normal distribution of the target data with right-sided tail creating a skew in the target. For example, we can see an interesting trend in in the histogram compared we first begin to notice the asymetric tail which corresponds to the fanning out of the data in the top right of the scatter plot for our models. 

Therefore we can hypothesize that this unequal (or better said, non-normal)distribution of sale prices has the potential to create this type of issue. In future works, we could use the target as the log of the sale prices to explore improvements to our model.

```python
def process_kaggle_data(kaggle_sub):    
    df_train = pd.read_csv('./datasets/train.csv')
    df_train.columns =[col.replace(' ', '_').lower() for col in df_train.columns]
    top_correlations = list(df_train.corr()[['saleprice']].sort_values('saleprice',ascending=False).head(10).index)[1:]
    # Automatically grabbing the top most correlated with features (with correlation > 0.5)
    df_subset = kaggle_sub.loc[:,top_correlations]
    df_subset.fillna(df_subset.median(),inplace = True)
    df_subset['neighborhood'] = kaggle_sub['neighborhood']
    
    df_subset_orig = df_train.loc[:,top_correlations]
    df_subset_orig.fillna(df_subset_orig.median(),inplace = True)
    df_subset_orig['neighborhood'] = df_train['neighborhood']

    df_orig_dummies = pd.get_dummies(df_subset_orig,columns = ['neighborhood'],drop_first=True)
    #df_orig_dummies.head()
    
    kaggle_dummies = pd.get_dummies(df_subset,columns = ['neighborhood'],drop_first=True)
    dummies_difference = [item for item in list(df_orig_dummies.columns) if item not in list(kaggle_dummies.columns)]
    
    kaggle_dummies[dummies_difference] = 0
    
    features = list(kaggle_dummies.columns)
    
    X = kaggle_dummies[features]
    #y=df_dummies['saleprice']
    return X

#X_kag = process_kaggle_data(kaggle_sub)

```
